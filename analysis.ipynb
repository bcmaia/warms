```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Load the CSV file into a DataFrame
df = pd.read_csv('fitness_test_21.csv')
df = df.iloc[:601]
```

```python
# Extract the 'best' column
best_values = df['best']

# Create a plot
plt.figure(figsize=(10, 6))
plt.plot(best_values, marker='o', linestyle='-')
plt.title('Best FItness Progression')
plt.xlabel('Time (1s)')
plt.ylabel('Fitness')
plt.grid(True)

# Show the plot
plt.show()
```
